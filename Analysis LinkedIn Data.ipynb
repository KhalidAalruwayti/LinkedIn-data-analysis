```python
def transform_duration(duration):
    if isinstance(duration, str):
        # extract year and month values using regular expression
        pattern = r'(?:(\d*\.?\d+)\s*(?:year|yr|y)(?:s|\b))?\s*(?:(\d*\.?\d+)\s*(?:month|mth|mo|mos|\b))?'  # updated regex pattern
        matches = re.findall(pattern, duration)
        duration_values = [(float(y or 0) + float(m or 0) / 12) for y, m in matches]

        # calculate total duration in years
        duration_float = sum(duration_values)

        return duration_float
    else:
        return None

# apply the transform_duration() function to the duration column
experience['duration'] = experience['duration'].apply(transform_duration)
job_class_list = {col: job_classification[col].dropna().tolist() for col in job_classification.columns}
experience['job_class'] = experience['position_name'].apply(lambda x: [col for col, lst in job_class_list.items() if x.lower() in [elem.lower() for elem in lst]])
experience['job_class'] = experience['job_class'].apply(lambda x: ', '.join(x))

```
